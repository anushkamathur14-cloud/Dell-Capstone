# Synthetic Environment Demo

Minimal walkthrough for world-spec driven generation and validation.

```python
from synthetic_env.world_spec.loader import load_world_spec
from synthetic_env.synthetic_population.generator import generate_population
from synthetic_env.treatment_space.generator import generate_experiment_tables
from synthetic_env.outcome_simulator.simulator import simulate_observations, summarize_metrics
from synthetic_env.validation.checks import create_validation_report
```

```python
spec = load_world_spec('configs/world_spec.yaml')
population = generate_population(spec, n_users=3000, seed=7)
experiments, arms = generate_experiment_tables(spec, experiment_id='exp_demo_001')
observations = simulate_observations(population, experiment_id='exp_demo_001', arm_ids=arms['arm_id'].tolist(), seed=7)
metrics = summarize_metrics(observations, experiment_id='exp_demo_001')

population.head(), arms.head(), metrics
```

```python
report = create_validation_report(population, experiments, arms, observations, metrics)
report
```
